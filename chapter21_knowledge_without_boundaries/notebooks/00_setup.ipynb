```python
print("Loading embedding model (first run downloads ~90MB)...")
model = SentenceTransformer('all-MiniLM-L6-v2')

# Test the model
test_sentences = [
    "The railroad was not liable for the plaintiff's injury.",
    "The court held that foreseeability determines the scope of duty."
]
test_embeddings = model.encode(test_sentences)
similarity = np.dot(test_embeddings[0], test_embeddings[1]) / (
    np.linalg.norm(test_embeddings[0]) * np.linalg.norm(test_embeddings[1])
)

print(f"  ✓ Model loaded: all-MiniLM-L6-v2")
print(f"  ✓ Embedding dimension: {test_embeddings.shape[1]}")
print(f"  ✓ Test similarity between two legal sentences: {similarity:.3f}")
print("Embedding model verified.")
```
